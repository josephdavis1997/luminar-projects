```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_PCA,y,test_size=0.2, random_state=30, stratify=y)
```

```python
from sklearn.svm import SVC  #SVC - Supoprt Vector Classifier
cls=SVC()
```

```python
from sklearn.model_selection import GridSearchCV
parameters = {'C':[0.001,0.1], 'gamma':[0.1],'kernel':['rbf']}

#C - Regularization - squared l2 error is calculated
#gamma - Fit of the model
#kernal - kernal to be used 
```

```python
grid=GridSearchCV(cls, parameters,cv=5)
grid.fit(X_train, y_train)
```

```python
#to get the best parameter for the model
grid.best_params_
```

```python
svclassifier=SVC(C=0.1,gamma=0.1,kernel='rbf',probability=True)
svclassifier.fit(X_train,y_train)
```

```python
ypred=svclassifier.predict(X_test)
```

```python
#accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_test,ypred)
```
